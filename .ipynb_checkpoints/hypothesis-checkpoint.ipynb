# Challenge 2 - Hypothesis Tests

In this section of the lab, we will test whether the hourly wage of all hourly workers is significantly different from $30/hr. Import the correct one sample test function from scipy and perform the hypothesis test for a 95% two sided confidence interval.

```python
# First we need an array for just the hourly rate and we might want to drop NaN values, because that belong to people who are no hourly workers

hourly_rate = df['Hourly Rate'].dropna()
hourly_rate
```

```python
from scipy.stats import ttest_1samp

#we select a value for alpha of 0.05 (p-value threshold, significance level)

# Two-sided test:
# Null hypothesis or H0: mean hourly wage = $30/h
# Alternative hyp or H1: mean hourly wage <> (unequal) $30/h

stat, pval = ttest_1samp(hourly_rate, 30)
print('stat is  ', stat)
print('pvalue for the two-tailed test is ', pval)
```

```python
# Because p is close to zero we reject H0. Because the stat is positive we can argue that the mean hourly wage might actually be higher than $30/h.
```

We are also curious about salaries in the police force. The chief of police in Chicago claimed in a press briefing that salaries this year are higher than last year's mean of $86000/year a year for all salaried employees. Test this one sided hypothesis using a 95% confidence interval.

Hint: A one tailed test has a p-value that is half of the two tailed p-value. If our Null hypothesis is less than or equal to, then to reject, the test statistic must be positive (rejects the <) and the p-value below 0.05 (rejects the =).

```python
# At first we need to get all the salary from the police force. Since there are some NaN values, I have to drop them first.

police_salary = df[df['Department'] == 'POLICE']['Annual Salary'].dropna()
police_salary
```

```python
#we select a value for alpha of 0.05 (p-value threshold, significance level)

# One-sided test:
# Null hypothesis or H0: mean anual salary <= $86000
# Alternative hyp or H1: mean anual salary > $86000

stat, pval = ttest_1samp(police_salary, 86000)
print('stat is  ', stat)
print('pvalue for the one-tailed test is ', pval/2)
```

```python
# Because p is close to zero and stat is positive we reject H0. Because the stat is positive we can argue that the mean salary is actually higher than $86000.
```

The workers from the department with the most hourly workers have complained that their hourly wage is less than $35/hour. Using a one sample t-test, test this one-sided hypothesis at the 95% confidence level.

```python
#First we select again the data. Here we need the hourly wage for the STREETS & SAN Department

workers_salary = df[df['Department'] == 'STREETS & SAN']['Hourly Rate'].dropna()
workers_salary
```

```python
# we select a value for alpha of 0.05 (p-value threshold, significance level)

# One-sided test:
# Null hypothesis or H0: mean hourly wage >= $35/hour
# Alternative hyp or H1: mean hourly wage < $35/hour

stat, pval = ttest_1samp(workers_salary, 35)
print('stat is  ', stat)
print('pvalue for the one-tailed test is ', pval/2)
```

```python
# P is close to zero but the stat is positive so we can not reject H0. A positive stat in this case indicates, that the mean is higher than 35.
```

# Challenge 3: To practice - Constructing Confidence Intervals

While testing our hypothesis is a great way to gather empirical evidence for accepting or rejecting the hypothesis, another way to gather evidence is by creating a confidence interval. A confidence interval gives us information about the true mean of the population. So for a 95% confidence interval, we are 95% sure that the mean of the population is within the confidence interval. 
).

To read more about confidence intervals, click [here](https://en.wikipedia.org/wiki/Confidence_interval).


In the cell below, we will construct a 95% confidence interval for the mean hourly wage of all hourly workers. 

The confidence interval is computed in SciPy using the `t.interval` function. You can read more about this function [here](https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.t.html).

To compute the confidence interval of the hourly wage, use the 0.95 for the confidence level, number of rows - 1 for degrees of freedom, the mean of the sample for the location parameter and the standard error for the scale. The standard error can be computed using [this](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.sem.html) function in SciPy.

```python
import scipy.stats

confidence_level = 0.95
degrees_freedom = len(hourly_rate) - 1
sample_mean = np.mean(hourly_rate)
sample_standard_error = scipy.stats.sem(hourly_rate)

confidence_interval = scipy.stats.t.interval(confidence_level, 
                                             degrees_freedom, 
                                             sample_mean, 
                                             sample_standard_error)

print( 'confidence interval is ', confidence_interval, '.' )
```

```python
# In the Hypothesis Test we rejected the H0 that the mean hourly rate was 30. 
# No we can see that we are 95% sure that the mean of the hourly rate is in the range from 34.30 to 34.87
```

Now construct the 95% confidence interval for all salaried employeed in the police in the cell below.

```python
confidence_level = 0.95
degrees_freedom = len(police_salary) - 1
sample_mean = np.mean(police_salary)
sample_standard_error = scipy.stats.sem(police_salary)

confidence_interval = scipy.stats.t.interval(confidence_level, 
                                             degrees_freedom, 
                                             sample_mean, 
                                             sample_standard_error)

print( 'confidence interval is ', confidence_interval, '.' )
```

```python
# In the Hypothesis Test we rejected the H0 that the mean yearly salary was below 86000$. 
# No we can see that we are 95% sure that the mean of the yearly salary is in the range from 88534.12 to 89134.12
```
